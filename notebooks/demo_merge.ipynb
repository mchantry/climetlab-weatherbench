```python
import climetlab as cml

acmlds = cml.load_dataset(
    "weatherbench",
    year=2018,
    parameter="temperature_850",
)
a = acmlds.to_xarray()

bcmlds = cml.load_dataset(
    "weatherbench",
    year=2018,
    parameter="geopotential_500",
)
b = bcmlds.to_xarray()

from climetlab.sources.multi import MultiSource
m = MultiSource([acmlds.source,bcmlds.source])

m.to_xarray(xarray_open_mfdataset_kwargs=dict(compat='override'))
#m.to_xarray(xarray_open_mfdataset_kwargs={'merge_kwargs': dict(compat='override')})
```
